# Problem 1

```python
def is_safe(report):
    if report[0] == report[1]:
        return 0

    decreasing = report[0] > report[1]
    diffs = [(-2*decreasing+1) * (report[i+1] - report[i]) for i in range(len(report)-1)]

    if all([x >= 1 for x in diffs]) and all([x <= 3 for x in diffs]):
        return 1
    else:
        return 0
        
res = sum([is_safe(x) for x in basedata])
res
```

# Problem 2

```python
def is_safe(report):
    def valid_diffs(diffs):
        return all([x >= 1 for x in diffs]) and all([x <= 3 for x in diffs])
    
    decreasing = report[0] > report[1]
    diffs = [(-2*decreasing+1) * (report[i+1] - report[i]) for i in range(len(report)-1)]

    if valid_diffs(diffs) and not report[0] == report[1]:
        return 1
    
    for i in range(len(report)):
        tmp_list = report[:i]+report[i+1:]
        
        decreasing = tmp_list[0] > tmp_list[1]
        diffs = [(-2*decreasing+1) * (tmp_list[i+1] - tmp_list[i]) for i in range(len(tmp_list)-1)]

        if valid_diffs(diffs):
            return 1

    return 0
        
res = sum([is_safe(x) for x in basedata])
res
```
